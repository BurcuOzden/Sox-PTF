```python
# load libraries
import seaborn as sns
from matplotlib import pyplot as plt
import pandas as pd
import shutil
import os
import random
import numpy as np
import matplotlib.pyplot as plt
import csv
from statistics import mean 
```

```python
a=pd.read_csv(r"dna_sox_equilibrated.prmsd", sep=" ")
a['Type']='Sox:DNA'
b=pd.read_csv(r"free_nuc_equilibrated.prmsd", sep=" ")
b['Type']='Free Nuc'
c=pd.read_csv(r"sox_shl2_equilibrated.prmsd", sep=" ")
c['Type']='Sox:SHL2'
d=pd.concat([a, b, c])
print(d)
```

```python
# RMSD analysis as violin plot - for shl2 and Sox:DNA
plt.style.use('default')
plt.style.use('ggplot')
plt.style.use('default')
plt.style.use('seaborn-white')
plt.figure(figsize=(4,5))
my_pal1=["#2e487c","#7b923c","#d95e46"]
#my_pal2=["#011959","#767634","#e19866"]
ax = sns.violinplot(x="Type", y="P-RMSD",palette=my_pal1, data=d)
plt.xticks(fontsize=14, fontweight='bold')
plt.yticks(fontsize=14, fontweight='bold')
plt.ylim(ymax = 5, ymin = 0)
plt.xlabel(' ', fontsize=14, fontweight='bold')
plt.ylabel('P-RMSD ($\AA$)', fontsize=14, fontweight='bold')
plt.xticks(rotation=90)
ax.grid(which='major', axis='y', linestyle='--')
plt.savefig("PRMSD-dna_freenuc_shl2.pdf",dpi=300,transparent=True,bbox_inches='tight')
```

```python
c=pd.read_csv(r"sox_shl2_equilibrated.prmsd", sep=" ")
c['Type']='Sox:SHL2'
e=pd.read_csv(r"sox_dyad_equilibrated.prmsd", sep=" ")
e['Type']='Sox:dyad'
f=pd.read_csv(r"sox_shl4_equilibrated.prmsd", sep=" ")
f['Type']='Sox:SHL4'
g=pd.concat([c, e, f])
print(g)
```

```python
# RMSD analysis as violin plot - for shl2, dyad and SHL4
plt.style.use('default')
plt.style.use('ggplot')
plt.style.use('default')
plt.style.use('seaborn-white')
plt.figure(figsize=(4,5))
my_pal1=["#d95e46","#e2ad3c","#7b923c"]
#my_pal2=["#011959","#767634","#e19866"]
ax = sns.violinplot(x="Type", y="P-RMSD",palette=my_pal1, data=g)
plt.xticks(fontsize=14, fontweight='bold')
plt.yticks(fontsize=14, fontweight='bold')
plt.ylim(ymax = 5, ymin = 0)
plt.xlabel(' ', fontsize=14, fontweight='bold')
plt.ylabel('P-RMSD ($\AA$)', fontsize=14, fontweight='bold')
plt.xticks(rotation=90)
ax.grid(which='major', axis='y', linestyle='--')
plt.savefig("PRMSD-shl2_dyad_shl4.pdf",dpi=300,transparent=True,bbox_inches='tight')
```
