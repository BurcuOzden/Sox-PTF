```python
#Â Please do not forget to change the path of file with yours!

dna_sox_sim1=pd.read_csv(r"/Users/klab01/Desktop/github/Raw_Data/RMSD/dna_sox11_sim1.rmsd-sorted", sep="\t")
dna_sox_sim2=pd.read_csv(r"/Users/klab01/Desktop/github/Raw_Data/RMSD/dna_sox11_sim2.rmsd-sorted", sep="\t")

freenuc_sim1=pd.read_csv(r"/Users/klab01/Desktop/github/Raw_Data/RMSD/601_SHL024_sim1.rmsd-sorted", sep="\t")
freenuc_sim2=pd.read_csv(r"/Users/klab01/Desktop/github/Raw_Data/RMSD/601_SHL024_sim2.rmsd-sorted", sep="\t")

sox_shl2_sim1=pd.read_csv(r"/Users/klab01/Desktop/github/Raw_Data/RMSD/Sox11_shl2_sim1.rmsd-sorted", sep="\t")
sox_shl2_sim2=pd.read_csv(r"/Users/klab01/Desktop/github/Raw_Data/RMSD/Sox11_shl2_sim2.rmsd-sorted", sep="\t")

sox_dyad_sim1=pd.read_csv(r"/Users/klab01/Desktop/github/Raw_Data/RMSD/Sox11_dyad_sim1.rmsd-sorted", sep="\t")
sox_dyad_sim2=pd.read_csv(r"/Users/klab01/Desktop/github/Raw_Data/RMSD/Sox11_dyad_sim2.rmsd-sorted", sep="\t")

sox_shl4_sim1=pd.read_csv(r"/Users/klab01/Desktop/github/Raw_Data/RMSD/Sox11_shl4_sim1.rmsd-sorted", sep="\t")
sox_shl4_sim2=pd.read_csv(r"/Users/klab01/Desktop/github/Raw_Data/RMSD/Sox11_shl4_sim2.rmsd-sorted", sep="\t")


# Arranges the raw P-RMSD files to draw violin plots.

dna_sox_sim1.columns = ["pdb_id", "prmsd"]
dna_sox_sim1.head()
dna_sox_sim1['Type']='Sox:DNA'
dna_sox_sim2.columns = ["pdb_id", "prmsd"]
dna_sox_sim2.head()
dna_sox_sim2['Type']='Sox:DNA'

freenuc_sim1.columns = ["pdb_id", "prmsd"]
freenuc_sim1.head()
freenuc_sim1['Type']='Free Nuc'
freenuc_sim2.columns = ["pdb_id", "prmsd"]
freenuc_sim2.head()
freenuc_sim2['Type']='Free Nuc'

sox_shl2_sim1.columns = ["pdb_id", "prmsd"]
sox_shl2_sim1.head()
sox_shl2_sim1['Type']='Sox:SHL2'
sox_shl2_sim2.columns = ["pdb_id", "prmsd"]
sox_shl2_sim2.head()
sox_shl2_sim2['Type']='Sox:SHL2'

sox_dyad_sim1.columns = ["pdb_id", "prmsd"]
sox_dyad_sim1.head()
sox_dyad_sim1['Type']='Sox:dyad'
sox_dyad_sim2.columns = ["pdb_id", "prmsd"]
sox_dyad_sim2.head()
sox_dyad_sim2['Type']='Sox:dyad'

sox_shl4_sim1.columns = ["pdb_id", "prmsd"]
sox_shl4_sim1.head()
sox_shl4_sim1['Type']='Sox:SHL4'
sox_shl4_sim2.columns = ["pdb_id", "prmsd"]
sox_shl4_sim2.head()
sox_shl4_sim2['Type']='Sox:SHL4'


data_dna_freenuc_shl2=pd.concat([dna_sox_sim1, dna_sox_sim2, freenuc_sim1, freenuc_sim2, sox_shl2_sim1, sox_shl2_sim2])

data_shl2_dyad_shl4=pd.concat([sox_shl2_sim1, sox_shl2_sim2, sox_dyad_sim1, sox_dyad_sim2, sox_shl4_sim1, sox_shl4_sim2])

```

```python
# RMSD analysis as violin plot - for Sox:DNA, Free Nucleosome and Sox11:SHL2 and Sox:DNA
plt.style.use('default')
plt.style.use('ggplot')
plt.style.use('default')
plt.style.use('seaborn-white')
plt.figure(figsize=(4,5))
my_pal1=["#2e487c","#7b923c","#d95e46"]
ax = sns.violinplot(x="Type", y="prmsd",palette=my_pal1, data=data_dna_freenuc_shl2)
plt.xticks(fontsize=14, fontweight='bold')
plt.yticks(fontsize=14, fontweight='bold')
plt.ylim(ymax = 5, ymin = 0)
plt.xlabel(' ', fontsize=14, fontweight='bold')
plt.ylabel('P-RMSD ($\AA$)', fontsize=14, fontweight='bold')
plt.xticks(rotation=90)
ax.grid(which='major', axis='y', linestyle='--')

# Since it saves a figure in the folder, be careful!
plt.savefig("PRMSD-dna_freenuc_shl2.pdf",dpi=300,transparent=True,bbox_inches='tight')
```
