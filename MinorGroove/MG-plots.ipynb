```python
# load libraries
import pandas as pd
import shutil
import os
import random
import matplotlib.pyplot as plt
import csv
from statistics import mean
import seaborn as sns
import numpy as np
```

```python
# This section calculates the mean, max and min from the appended minor groove file. 
# The generated file is used to plot min and max values at corresponding DNA index.
# do not forget to change the range (for i in range(4,141):) according to your minow groove indexes.
# Be careful before running, since it generates new file!

with open ("sox_shl2.minor", "r") as amber:
    amberparameters=amber.readlines()
for i in range(4,141):
    minor_amber=[]

    for j in range(0,len(amberparameters)):
        if amberparameters[j].split(' ')[0] == str(i):
            minor_amber.append(float(amberparameters[j].split(' ')[1]))
    #print(minor_amber)
            
    with open ("sox_shl2_min_max.csv", "a") as max_min:
        max_min.write(str(i) + ';' + str(max(minor_amber)) + ';' + str(min(minor_amber)) + ';' + str(mean(minor_amber)) + '\n')
```

```python
# Minor groove plot of free nucleosome complex 
d=pd.read_csv(r"free_nuc_min_max.csv", sep=";")
e=pd.read_csv(r"cryoem_shl2_min_max_only.minor", sep="\t")
f=pd.DataFrame(d)
max_a=f['minor_max']
min_a=f['minor_min']
index_a=f['index']

plt.figure(figsize=(9,5))
sns.lineplot(x="index", y="minor_max", linewidth=2, color='#7b923c', data=d) #max
facet=sns.lineplot(x="index", y="minor_min", linewidth=2, color='#7b923c', data=d) #min
plt.fill_between(index_a,max_a,min_a, color='#7b923c', alpha=0.6)
sns.lineplot(x="resi", y="minor", color='gray', linewidth=4, markers=True, data=e)
plt.ylim(ymax = 25, ymin = 0)
plt.xlim(xmax=140, xmin=5)
plt.xticks(fontsize=14, fontweight='bold')
plt.yticks(fontsize=14, fontweight='bold')
plt.xlabel('Free Nuc', fontsize=14, fontweight='bold')
plt.ylabel('MG Width ($\AA$)', fontsize=14, fontweight='bold')
ax=plt.gca()
ax.set_xticklabels([])
plt.grid(linestyle='--')
plt.grid(True)
plt.savefig("MG-free_nuc-equilibrated.pdf",dpi=300,bbox_inches='tight')
plt.show()
```
